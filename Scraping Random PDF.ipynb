```python
import tabula
import pandas as pd
import re

complete_result = {}

# Read the PDF file and extract the table
def regionalResult(doc):
    df_list = tabula.io.read_pdf(doc, pages='all')
    for i, element in enumerate(df_list):
        compile(df_list[i])

def compile(competition):  
    ref = -1
    div = ''
    column = ['Technical_Evaluation','Execution','Control','Trick_Diversity','Space_Use','Choreography','Construction','Body_Control',
              'Showmanship','Freestyle_Evaluation','Evaluation_Score','Stop','Discard','Cut','Final_Score','Placing']
    for index, title in competition['CloudNative Inc. Presents'].iteritems():
        if(type(title) == str and '2023' in title): 
            ref = index
            div = title
            break
            
    print(div)
    database = {}
    for i in range(ref+2,len(competition)):
        if 'Unnamed: 10' not in competition:
            name = re.sub("^\d+ ", "", competition.iloc[i,0])
            add = 0
        else :
            name = competition.iloc[i,1]
            add = 1
        
        if name not in database:
            indiv_result = {}
            count = 1 + add;
            subCount = 0;
            for indCol in column:
                if(count != 12 + add):
                    indiv_result[indCol] = competition.iloc[i,count]
                    count += 1
                elif count == 12 + add and subCount < 2:
                    indiv_result[indCol] = competition.iloc[i,count].split()[subCount]
                    subCount += 1
                elif count == 12 + add and subCount == 2:
                    indiv_result[indCol] = competition.iloc[i,count].split()[subCount]
                    count += 1
            database[name] = indiv_result
    complete_result[div] = database
    
def sorting(division):
    sorted_division = dict(sorted(division.items(), key=lambda x: float(x[1]['Final_Score']), reverse=True))

    placing = 1
    for competitor, values in sorted_division.items():
        values['Placing'] = str(placing)
        placing += 1
    
    return sorted_division


regionalResult('2023ej-result-final-en.pdf')
regionalResult('2023wj-result-final-en.pdf')
regionalResult('2023sj-result-final-en.pdf')
regionalResult('2023cj-result-final-en.pdf')

results = {
    "1A": {},
    "2A": {},
    "3A": {},
    "4A": {},
    "5A": {}
}

for division, competitors in complete_result.items():
    if re.search("1A", division):
        for key, value in competitors.items():
            results['1A'][key] = value
    elif re.search("2A", division):
        for key, value in competitors.items():
            results['2A'][key] = value
    elif re.search("3A", division):
        for key, value in competitors.items():
            results['3A'][key] = value
    elif re.search("4A", division):
        for key, value in competitors.items():
            results['4A'][key] = value
    elif re.search("5A", division):
        for key, value in competitors.items():
            results['5A'][key] = value

sorted1A = sorting(results['1A'])
sorted2A = sorting(results['2A'])
sorted3A = sorting(results['3A'])
sorted4A = sorting(results['4A'])
sorted5A = sorting(results['5A'])
```
